```python
import pandas as pd


# Charger les données de pollution transformées
df_pollution = pd.read_csv(r'C:\Users\Dora\Desktop\Projet AI\data\raw\pollution_data_transformed.csv')

# Données démographiques
demographic_data = {
    'City': ['Los Angeles', 'Paris', 'Tokyo', 'Antananarivo', 'Nairobi', 'Lima'],
    'Population': [10039107, 2140526, 13929286, 1391433, 4397073, 9674755],
    'Density': [3276, 21383, 6169, 3097, 6000, 3220],
    'Urbanization': [89, 98, 100, 69, 61, 81],
    'Average Income': [60000, 45000, 55000, 1000, 1500, 4500],
    'Education Level': [32, 29, 37, 10, 12, 15]
}

# Données géographiques
geographic_data = {
    'City': ['Los Angeles', 'Paris', 'Tokyo', 'Antananarivo', 'Nairobi', 'Lima'],
    'Altitude': [89, 35, 40, 1276, 1795, 1540],
    'Proximity to Industry': [5.0, 3.0, 2.0, 0.5, 1.0, 1.5]
}

# Conversion en DataFrames
df_demographic = pd.DataFrame(demographic_data)
df_geographic = pd.DataFrame(geographic_data)

# Fusion des trois DataFrames sur la colonne 'City'
df_combined = df_pollution.merge(df_demographic, on='City').merge(df_geographic, on='City')

# Afficher les données combinées
#print(df_combined.head())


# Sélectionner uniquement les colonnes numériques
numeric_columns = df_combined.select_dtypes(include='number')

# Calcul des corrélations
correlation_matrix = numeric_columns.corr()

# Affichage de la matrice de corrélation
print(correlation_matrix)


```
